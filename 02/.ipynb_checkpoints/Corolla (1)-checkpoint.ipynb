```python
import matplotlib.pyplot as plt
X = df.iloc[:,1:]  #independent columns
y = df.iloc[:,0]    #target column i.e price range
#get correlations of each features in dataset
corrmat = df.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(20,20))
#plot heat map
g=sns.heatmap(df[top_corr_features].corr(),annot=True,cmap="RdYlGn")
```

```python
y=df['Price']
x1=df['Age_08_04']
x2=df['KM']
x3=df['cc']
x4=df['Doors']
x5=df['Gears']
x6=df['Weight']
```

```python
import statsmodels.formula.api as smf
model=smf.ols("y~x1+x2+x3+x4+x5+x6",data=df).fit()
```

```python
model.summary()
```

```python
newdata=pd.DataFrame({'x1':21,'x2':41711,'x3':1300,'x4':3,'x5':5,'x6':1247},index=[1])
model.predict(newdata)
```

```python
rsq_x1 = smf.ols('x1~x2+x3+x4+x5+x6',data=df).fit().rsquared  
vif_x1 = 1/(1-rsq_x1) # 16.33

rsq_x2 = smf.ols('x2~x1+x3+x4+x5+x6',data=df).fit().rsquared  
vif_x2 = 1/(1-rsq_x2) # 564.98

rsq_x3 = smf.ols('x3~x2+x1+x4+x5+x6',data=df).fit().rsquared  
vif_x3 = 1/(1-rsq_x3) #  564.84

rsq_x4 = smf.ols('x4~x2+x3+x1+x5+x6',data=df).fit().rsquared  
vif_x4 = 1/(1-rsq_x4) #  16.35

rsq_x5 = smf.ols('x4~x2+x3+x1+x5+x6',data=df).fit().rsquared  
vif_x5 = 1/(1-rsq_x5) #  16.35

rsq_x6 = smf.ols('x4~x2+x3+x1+x5+x6',data=df).fit().rsquared  
vif_x6 = 1/(1-rsq_x6) #  16.35

# Storing vif values in a data frame
d1 = {'Variables':['x1','x2','x3','x4','x5','x6'],'VIF':[vif_x1,vif_x2,vif_x3,vif_x4,vif_x5,vif_x6]}
Vif_frame = pd.DataFrame(d1)  
Vif_frame
```

```python
import statsmodels.api as sm
qqplot=sm.qqplot(model.resid,line='q') # line = 45 to draw the diagnoal line
plt.title("Normal Q-Q plot of residuals")
plt.show()
```

```python
list(np.where(model.resid>5000))
```

```python
list(np.where(model.resid<-5000))
```

```python
model_influence = model.get_influence()
(c, _) = model_influence.cooks_distance
#Plot the influencers values using stem plot
fig = plt.subplots(figsize=(20, 7))
plt.stem(np.arange(len(df)), np.round(c, 3))
plt.xlabel('Row index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
(np.argmax(c),np.max(c))
```
