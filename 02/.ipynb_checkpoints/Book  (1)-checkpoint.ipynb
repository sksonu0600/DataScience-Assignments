```python
df1=pd.DataFrame(book)
df2=df1.pivot_table(index='User.ID',columns='Book.Title',values='Book.Rating',aggfunc='mean')
```

```python
df2
```

```python
df2.index = (book['User.ID'].unique())
```

```python
df2
```

```python
df2.fillna(0,inplace=True)
df2
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python

df3= 1 - pairwise_distances( df2.values,metric='cosine')
```

```python
df3
```

```python
df4 = pd.DataFrame(df3)
```

```python

df4.index=(book['User.ID'].unique())
df4.columns=(book['User.ID'].unique())
```
